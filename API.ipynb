```python
# Importando do Kit de desenvolvimento de software Python (SDK)
import google.generativeai as genai
# Protocolo padrão para garantir a segurança da chave API
from google.colab import userdata
# Importação do pandas para eventual leitura de dataframes
import pandas as pd
# Configurando a API com sua chave
GOOGLE_API_KEY=userdata.get('teste')
genai.configure(api_key=GOOGLE_API_KEY)

# Instalação do pacote Pillow para reconhecimento de imagens
!pip install pillow
```

```python
# Seleção de qual modelo generativo iremos usar - neste caso: Gemini-pro (textos)
model = genai.GenerativeModel('gemini-pro')
# Na área prompt é fornecida a entrada do usuário, o que será requisitado da IA
prompt = """
Explique o que é uma IDE e como funciona de forma resumida

"""
# A resposta da IA é armazenada na variável response
response = model.generate_content(prompt)
# Impressão da resposta
print(response.text)
```

```python
df = pd.read_csv('/content/random - Página1.csv')

response = model.generate_content(f"Leia os valores da variável {df} importado para o notebook")
print(response.text)
```

```python
# Aqui é importada a função PIL, possibilitando o import da imagem
import PIL.Image
# Aqui é aberta a imagem e armazenada na variável img
img = PIL.Image.open(r'/content/exemplo-email.jpg')
# visualização da imagem
img
```

```python
# Aqui usa-se outro modelo do Gemini, Gemini-pro-vision, o modelo generativo voltado para reconhecimento de imagens, faremos uma requisição juntando o prompt com a imagem do upload
model = genai.GenerativeModel('gemini-pro-vision')
prompt = """
Reescreva o email com outras palavras, sem perder o contexto principal, porém mais formal

"""
response = model.generate_content([prompt, img])
```

```python
# Apresentando a resposta no formato de texto
print(response.text)
```
